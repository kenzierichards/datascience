## visualizing Titanic dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
#plotting fare against age - most passengers paid under $100 across all ages

sns.jointplot(x="fare", y="age", data=titanic)
```

```python
#most people paid the lowest fare to get on

sns.distplot(titanic['fare'], kde=False, color='red', bins=30)
```

```python
#first class tended to be older people w/ more variance, third younger people

sns.boxplot(x='class', y='age', data=titanic)
```

```python
sns.swarmplot(x='class', y='age', data=titanic, palette='rainbow')
```

```python
#more male passengers
sns.countplot(x='sex', data=titanic)
```

```python
#alone & adult_male are highly correlated
sns.heatmap(titanic.corr(), cmap='coolwarm')
```

```python
#facet grid - two graphs, separated by a specific column
#distribution of age among sexes - men were older, women were younger

g = sns.FacetGrid(data=titanic, col='sex')
g.map(plt.hist, 'age')
```
