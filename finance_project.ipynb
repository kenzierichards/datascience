## analyzing stock data

```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
%matplotlib inline
```

```python
#google API didn't work
df = pd.read_pickle('all_banks') 
df
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
#already put together
bank_stocks = df
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

```python
#get cross section of the close key at the stock info level
#Close is a index of the Stock Info subindex

bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

```python
#for each ticker
#set a new column to the bank code and return
#perform function on the ticker's close number

returns = pd.DataFrame()
for ticker in tickers:
    returns[ticker+" Return"] = bank_stocks[ticker]['Close'].pct_change()
    
returns
```

```python
p = sns.pairplot(returns)
p

#citigroup returns are steep
```
