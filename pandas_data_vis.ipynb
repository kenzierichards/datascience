```python
#visualizing a data frame made up of random data

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df3 = pd.read_csv('df3')
```

```python
df3.info()
```

```python
df3.head()
```

```python
df3.plot.scatter(x='b', y='a', figsize=(12, 3))
```

```python
df3['a'].plot.hist()
```

```python
#improve style

plt.style.use('ggplot') #import style sheet
df3['a'].plot.hist(bins=20, alpha=.5)
```

```python
#create box plot from each column
df3[['a', 'b']].plot.box()
```

```python
#change line width
df3['d'].plot.kde(lw=5)
```

```python
#using loc to grab & plot a slice of indices

df3.loc[0:30].plot.area()
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```
